```python
import pandas as pd

xlsx_file_path = 'C:\\Users\\Priyanshu Kumar\\Documents\\xlsx\\Assignment_Timecard.xlsx'

df = pd.read_excel(xlsx_file_path)

# Sort the DataFrame by employee and date
df.sort_values(by=['Employee Name', 'Pay Cycle Start Date'], inplace=True)
df.reset_index(drop=True, inplace=True)

# Initialize variables to track consecutive days and previous shift end time
consecutive_days = 0
prev_shift_end = None

# Iterate over the records in the DataFrame
for index, row in df.iterrows():
    current_date = row['Pay Cycle Start Date']
    current_shift_start = row['Time']
    current_shift_end = row['Time Out']
    employee_name = row['Employee Name']
    position = row['Position ID']

    # Check for consecutive days
    if prev_shift_end is None or (current_date - prev_shift_end).days == 1:
        consecutive_days += 1
    else:
        consecutive_days = 0

    # Check for less than 10 hours between shifts but greater than 1 hour
    if prev_shift_end is not None:
        time_difference = (current_shift_start - prev_shift_end).seconds // 3600
        if 1 < time_difference < 10:
            print(f"{employee_name}, {position} has less than 10 hours between shifts on {current_date}")
            
            print( "less than 10 hours between shifts but greater than 1 hour")
    # Check for working more than 14 hours in a single shift
    shift_duration = (current_shift_end - current_shift_start).seconds // 3600
    if shift_duration > 14:
        print(f"{employee_name}, {position} worked more than 14 hours on {current_date}")

    # (Reset consecutive days count if needed)
    if consecutive_days == 7:
        print(f"{employee_name}, {position} worked for 7 consecutive days until {current_date}")
        consecutive_days = 0

    # (Update previous shift end time)
    prev_shift_end = current_shift_end
```
